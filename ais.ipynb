```python
# Definisikan fungsi keanggotaan
def trapezoidal_function(x, a, b, c, d):
    if x <= a or x >= d:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1
    elif c < x < d:
        return (d - x) / (d - c)
```

```python
#menentukan parameter fungsi Keanggotaan
#kelembaban  #a,b,c,d
kering      = [0,0,20,30]        
lembab      = [20,30,50,60]
basah       = [50,60,80,100]
#ammonia
normal      = [0,0,25,45]
sedang      = [25,45,65,85]
tinggi      = [65,85,100,100]

# Definisikan rentang nilai x
x = np.linspace(0, 100,1000)

###kelembaban
# Definisikan parameter fungsi keanggotaan untuk kategori kelembaban
a_kering, b_kering, c_kering, d_kering = 0, 0, 20, 30
y_kering = [trapezoidal_function(i, a_kering, b_kering, c_kering, d_kering) for i in x]

a_lembab, b_lembab, c_lembab, d_lembab = 20, 30, 50, 60
y_lembab = [trapezoidal_function(i, a_lembab, b_lembab, c_lembab, d_lembab) for i in x]

a_basah, b_basah, c_basah, d_basah = 50, 60, 80, 100
y_basah = [trapezoidal_function(i, a_basah, b_basah, c_basah, d_basah) for i in x]

# Plot fungsi keanggotaan untuk kategori kelembaban
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y_kering, label='Kering')
plt.plot(x, y_lembab, label='Lembab')
plt.plot(x, y_basah, label='Basah')
plt.title('Grafik Fungsi Keanggotaan Kelembaban')
plt.xlabel('Nilai Input')
plt.ylabel('Nilai Keanggotaan')
plt.legend()
plt.grid(True)

###ammonia
# Definisikan parameter fungsi keanggotaan untuk kategori ammonia
a_normal, b_normal, c_normal, d_normal = 0, 0, 25, 45
y_normal = [trapezoidal_function(i, a_normal, b_normal, c_normal, d_normal) for i in x]

a_sedang, b_sedang, c_sedang, d_sedang = 25, 45, 65, 85
y_sedang = [trapezoidal_function(i, a_sedang, b_sedang, c_sedang, d_sedang) for i in x]

a_tinggi, b_tinggi, c_tinggi, d_tinggi = 65, 85, 100, 100
y_tinggi = [trapezoidal_function(i, a_tinggi, b_tinggi, c_tinggi, d_tinggi) for i in x]

# Plot fungsi keanggotaan untuk kategori ammonia
plt.subplot(1, 2, 2)
plt.plot(x, y_normal, label='Normal')
plt.plot(x, y_sedang, label='Sedang')
plt.plot(x, y_tinggi, label='Tinggi')
plt.title('Grafik Fungsi Keanggotaan Ammonia')
plt.xlabel('Nilai Input')
plt.ylabel('Nilai Keanggotaan')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
